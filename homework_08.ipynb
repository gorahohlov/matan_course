### 1. Найти и изобразить области определения функций:

##### 1) $$z=\sqrt{(y \cdot \sin x)};$$

Подкоренное выражение не может быть меньше нуля. Стало быть $(y \cdot \sin x) \geq 0$ z будет определа в тех областях плосткости Oxy, где у $\sin x$ и у $y$ будут совпадать знаки.

$D(z): \Big(x\in\big[2\pi n;\pi (2n+1)\big]\wedge y\in\big[0;+\infty\big)\Big) \cup \Big(x\in 
\big[\pi (2n-1);2n\pi\big] \wedge y \in \big[0;-\infty\big)\Big), n \in \mathbb{N}$

```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
import math
warnings.filterwarnings('ignore')
x = np.linspace(-2 * math.pi, 4 * math.pi, 60)
y = np.linspace(-30, 30, 60)
X, Y = np.meshgrid(x, y)
Z = np.sqrt(Y * np.sin(X))
fig = plt.figure()
fig = plt.figure(figsize=(20, 18))
ax = fig.add_subplot(projection='3d', elev=90, azim=-90)
ax.plot_surface(X, Y, Z)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

### 2. Вычислить пределы:

##### 1) $$\lim_{x\to0\, y\to 0}(x+y) \cdot \sin\frac{1}{x}\cdot \cos\frac{1}{y};$$

Синус и косинус - функции ограниченные; функция $(x+y)$ стремиться к нулю - т.е. бесконечно малая. Произведение б.м. функции на ограниченные - есть бесконечно малая функция;
$$\lim_{x\to0\, y\to 0}(x+y) \cdot \sin\frac{1}{x}\cdot \cos\frac{1}{y}=0;$$

##### 2) $$\lim_{x\to1\,y\to2}\frac{2(x-1)(y-2)}{(x-1)^2+(y-2)^2};$$

Выполним замену $$\Big|(x-1)=u, (y-2)=v, (x\to1, u\to0), (y\to2, v\to0)\Big|$$
имеем, $\lim_{u,v\to0} \frac{2uv}{u^2+v^2}$ воспользуемся неравенством Коши: $u^2+v^2\geq 2uv$

$$\frac{2uv}{u^2+v^2}\leq\frac{2uv}{2uv}=1$$
значит $\frac{2uv}{u^2+v^2}$ тем более стремиться к 1;

Ответ: $$\lim_{x\to1\,y\to2}\frac{2(x-1)(y-2)}{(x-1)^2+(y-2)^2}=1;$$

##### 3) $$\lim_{x,y\to\infty}\Big(\frac{xy}{x^2+y^2}\Big)^{x^2};$$

Сделаем замену переменных чтобы перейти от бесконечности к стремлению к нулю:
$$\Big|\quad x=\frac{1}{u},\: y=\frac{1}{v},\: x,y \to \infty \Rightarrow u,v \to 0 \quad \Big|$$

$$\lim_{x,y\to\infty}\Big(\frac{xy}{x^2+y^2}\Big)^{x^2}=\lim_{u,v\to0}\ \Big(\frac{uv}{u^2+v^2}\Big)^{\frac{1}{u^2}}$$

Вновь поспользуемся неравенством Коши: $u^2+v^2 \geq 2uv$,
$$\frac{uv}{u^2+v^2}\leq \frac{uv}{2uv}=\frac{1}{2}, \quad \lim\Big(\frac{1}{2}\Big)^{\infty}\to0$$
стало быть выражение: $$0 \leq \frac{uv}{u^2+v^2} \leq \frac{1}{2
}$$ тем более стремиться к нулю, поскольку оно менее или раное $\frac{1}{2};$

Ответ: $$\lim_{x,y\to\infty}\Big(\frac{xy}{x^2+y^2}\Big)^{x^2}=0$$

##### 4) $$\lim_{x,y\to0}(x^2+y^2)\sin^3\bigg(\frac{1}{xy}\bigg);$$

множитель: $$\sin^3\bigg(\frac{1}{xy}\bigg)$$ являетя ограниченной функцией, множитель $(x^2+y^2)$ стремиться к нулю. Произведение бесконечно малой функции на ограниченную есть бесконечно малая функция. Ответ:
 $$\lim_{x,y\to0}(x^2+y^2)\sin^3\bigg(\frac{1}{xy}\bigg)=0$$

##### 5) $$\lim_{x,y\to0}\frac{x^2y^3}{x^2+y^2};$$

Сделаем замену $$\Big| \:x=r\cos\varphi, y=r\sin\varphi, x,y\to0: r\to0 \:\Big|$$

 $$\lim_{x,y\to0}\frac{x^2y^3}{x^2+y^2}=\lim_{r\to0} \frac{r^2\cos^2\varphi \cdot r^3\sin^3\varphi}{r^2 \cos^2 \varphi+r^2\sin^2 \varphi}=\lim_{r\to0} \frac{r^5\sin^2 2\varphi \cdot \sin\varphi}{4r^2}=\lim_{r\to0}\frac{r^3 \sin^2 2\varphi \sin\varphi}{4}=\frac{0}{4}=0;$$

проверим с помощью неравенства Коши: $x^2+y^2\geq 2xy$, значит $$\frac{x^2y^3}{x^2+y^2}\leq \frac{x^2\cdot y^3}{2xy}=\frac{xy^2}{2}; \Rightarrow \lim_{x,y\to0} \frac{xy^2}{2}=\frac{0}{2}=0;$$
Ответ: $$\lim_{x,y\to0}\frac{x^2y^3}{x^2+y^2}=0;$$
